### 1. Collect data from the Frankfurt Stock Exchange for AFX_X for 2017

```python
# 1. Collect data from Frankfurt Stock Exchange for AFX_X for 2017
r = requests.get('https://www.quandl.com/api/v3/datasets/FSE/AFX_X?start_date=2017-01-01&end_date=2017-12-31&api_key=MY_API_KEY')
```

### 2. Convert the returned JSON object into a Python dictionary

```python
# 2. Convert the returned JSON object into a Python dictionary
data = dict(r.json())
```

```python
# The 'data' key in 'dataset' contains all of the information, data will be assigned to just this list
data = data['dataset']['data']
```

According to the Quandl documentation, the data values structured as follows:      
    Date  
    Open  
    High  
    Low  
    Close  
    Change  
    Traded Volume  
    Turnover  
    Last Price of the Day  
    Daily Traded Units  
    Daily Turnover  

### 3. Calculate the highest and lowest opening prices for the stock in this period

```python
# 3. Calculate the highest and lowest opening prices for the stock in this period
# Create list containing opening prices, dropping null values
oprice = []
for i in range(0, len(data)):
    if data[i][1] == None:
        continue
    else:
        oprice.append(data[i][1])
```

```python
# Print high and low opening prices
print('Highest opening price of 2017 was ' + str(max(oprice)))
print('Lowest opening price of 2017 was ' + str(min(oprice)))
```

### 4. Calculate the largest change in any one day (based on High and Low price)

```python
# 4. Calculate the largest change in any one day (based on High and Low price)
# Create list containing difference between daily high and daily low, dropping null values
chprice = []
for i in range(0, len(data)):
    if data[i][2] == None or data[i][3] == None:
        continue
    else:
        chprice.append(data[i][2] - data[i][3])
```

```python
# Print largest change in any one day
print('Largest change in any one day was ' + str(round(max(chprice), 2)))
```

### 5. Calculate the largest change between any two days (based on Closing price)

```python
# 5. Calculate the largest change between any two days (based on Closing price)
# Create list containing difference between closing prices from one day to the next, dropping null values
chprice2 = []
for i in range(0, len(data)-1):
    if data[i][4] == None or data[i + 1][4] == None:
        continue
    else:
        chprice2.append(data[i + 1][4] - data[i][4])
```

```python
# Print largest loss and largest gain from one day to next, to identify which was greater in magnitude
print('Largest loss from one day to the next was ' + str(round(min(chprice2), 2)))
print('Largest gain from one day to the next was ' + str(round(max(chprice2), 2)))
```

```python
# Print largest change from one day to the next
print('Largest change from one day to the next was ' + str(round(max(chprice2), 2)))
```

### 6. Calculate the average daily trading volume during the year

```python
# 6. Calculate the average daily trading volume during the year
# Create list containing daily trading volume, dropping null values
tvolume = []
for i in range(0, len(data)):
    if data[i][6] == None:
        continue
    else:
        tvolume.append(data[i][6])
```

```python
# Calculate and print average daily trading volume
avgtvolume = sum(tvolume) / len(tvolume)
print('Average daily trading volume was ' + str(round(avgtvolume)))
```

### 7. Calculate the median daily trading volume during the year

```python
# 7. Calculate the median daily trading volume during the year
# Sort tvolume
tvolume.sort

# Calculate median and print result
if len(tvolume) % 2 == 0:
    tvolmedian = tvolume[(len(tvolume)/2)]
else:
    mid = len(tvolume)//2
    tvolmedian = (tvolume[mid - 1] + tvolume[mid]) / 2
print('The median daily trading volume was ' + str(round(tvolmedian)))
```
