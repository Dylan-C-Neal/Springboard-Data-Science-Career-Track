**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
# Subset data into two sets: one for 1998 and one for 2018
p1998 = prop_melt.loc[prop_melt['Year'] == 1998]
p2018 = prop_melt.loc[prop_melt['Year'] == 2018]
```

```python
# Extract average housing price for each borough and log in separate dictionaries for 1998 and 2018
dic1998 = {}
dic2018 = {}

for borough in p1998.index.unique():
    dic1998[borough] = p1998.loc[borough]['Average Housing Price'].mean()
    
for borough in p2018.index.unique():
    dic2018[borough] = p2018.loc[borough]['Average Housing Price'].mean()
```

```python
# Combine dictionaries into a DateFrame
avgprice = pd.DataFrame({'1998 Price Average':pd.Series(dic1998), '2018 Price Average':pd.Series(dic2018)})
```

```python
# Calculate price ratio of 1998 average price / 2018 average price and store in new column
avgprice['Price Ratio'] = avgprice['1998 Price Average'].divide(avgprice['2018 Price Average'])
```

```python
# Calculate percent growth from 1998 to 2018 and store in new column
avgprice['Percent Growth'] = 1/avgprice['Price Ratio'] * 100
avgprice['Percent Growth'] = avgprice['Percent Growth'].round(2)
```

```python
# print DataFrame
avgprice
```

```python
# Describe Percent Growth column
avgprice['Percent Growth'].describe()
```

```python
# Create a box plot of Percent Growth
avgprice['Percent Growth'].plot(kind='box', title='Percent Growth in Housing Price from 1998 to 2018 of London Boroughs')
```

```python
# Find which borough experienced the highest average growth in price from 1998 to 2018
avgprice.loc[avgprice['Percent Growth'] == avgprice['Percent Growth'].max()]
```

```python
# Find which borough experienced the lowest average growth in price from 1998 to 2018
avgprice.loc[avgprice['Percent Growth'] == avgprice['Percent Growth'].min()]
```

```python
# Drop outlier and reanalyze
avgprice2 = avgprice.drop('Hackney: E09000012')
```

```python
avgprice2['Percent Growth'].describe()
```

```python
avgprice2['Percent Growth'].plot(kind='box', title='Percent Growth in Housing Price from 1998 to 2018 of London Boroughs without Outlier')
```

```python
# After removing outlier, find which borough experienced the highest average growth in price from 1998 to 2018
avgprice2.loc[avgprice2['Percent Growth'] == avgprice2['Percent Growth'].max()]
```
